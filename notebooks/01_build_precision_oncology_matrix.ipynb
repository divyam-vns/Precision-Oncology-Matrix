```python
import pandas as pd
```

```python
approved = pd.read_csv("approved_targets.csv")
emerging = pd.read_csv("emerging_targets.csv")
resistance = pd.read_csv("resistance_mechanisms.csv")
```

```python
approved["Category"] = "FDA_Approved"
emerging["Category"] = "Emerging"
resistance["Category"] = "Resistance"
```

```python
approved = approved.rename(columns={
    "Gene": "Gene",
    "Variant": "Variant",
    "Cancer_Type": "Cancer_Type",
    "Drug": "Therapy",
    "Pathway": "Pathway"
})
```

```python
emerging = emerging.rename(columns={
    "Target": "Gene",
    "Drug": "Therapy",
    "Cancer_Type": "Cancer_Type",
    "Pathway": "Pathway"
})
```

```python
resistance = resistance.rename(columns={
    "Primary_Target": "Gene",
    "Initial_Therapy": "Therapy",
    "Tumor_Type": "Cancer_Type",
    "Resistance_Alteration": "Variant",
    "Pathway": "Pathway"
})
```

```python
master_matrix = pd.concat([approved, emerging, resistance], ignore_index=True)

master_matrix.head()
```

```python
master_matrix.to_csv("precision_oncology_master_matrix.csv", index=False)

master_matrix.to_json(
    "precision_oncology_master_matrix.json",
    orient="records",
    indent=4
)

print("MASTER MATRIX CREATED SUCCESSFULLY")
```

```python
patient_variants = [
    {
        "Gene": "KRAS",
        "Variant": "G12C",
        "Cancer_Type": "NSCLC"
    },
    {
        "Gene": "EGFR",
        "Variant": "L858R",
        "Cancer_Type": "NSCLC"
    },
    {
        "Gene": "ESR1",
        "Variant": "Y537S",
        "Cancer_Type": "Breast Cancer"
    }
]
```

```python
import pandas as pd

patient_df = pd.DataFrame(patient_variants)

patient_df
```

```python
trial_database = [
    {
        "Gene": "KRAS",
        "Variant": "G12C",
        "Drug": "Adagrasib",
        "Trial_ID": "NCT04685135",
        "Phase": "Phase II",
        "Cancer_Type": "NSCLC",
        "Mechanism": "KRAS G12C inhibitor"
    },
    {
        "Gene": "EGFR",
        "Variant": "L858R",
        "Drug": "Patritumab Deruxtecan",
        "Trial_ID": "NCT04619004",
        "Phase": "Phase II",
        "Cancer_Type": "NSCLC",
        "Mechanism": "HER3-directed ADC"
    },
    {
        "Gene": "ESR1",
        "Variant": "Y537S",
        "Drug": "Elacestrant",
        "Trial_ID": "NCT03778931",
        "Phase": "Phase III",
        "Cancer_Type": "Breast Cancer",
        "Mechanism": "Selective estrogen receptor degrader"
    }
]
```

```python
trial_df = pd.DataFrame(trial_database)

trial_df
```

```python
matched_trials = pd.merge(
    patient_df,
    trial_df,
    on=["Gene", "Variant", "Cancer_Type"],
    how="inner"
)

matched_trials
```

```python
matched_trials.to_json(
    "matched_trials_report.json",
    orient="records",
    indent=4
)

print("Clinical trial matching report generated.")
```
